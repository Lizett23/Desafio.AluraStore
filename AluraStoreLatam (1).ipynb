### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Ingreso total por cada tienda

```python


ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)

ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ['tienda', 'tienda2', 'tienda3', 'tienda4']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='m')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingreso ($)')
plt.xlabel('Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
ventas = pd.concat([tienda, tienda2, tienda3, tienda4])

plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=ventas,
    x='lon',
    y='lat',
    fill=True,
    cmap='Reds',
    thresh=0.05
)
plt.title('Mapa de Calor de Ventas Totales')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_t1 = tienda['Producto'].value_counts()
ventas_t2 = tienda2['Producto'].value_counts()
ventas_t3 = tienda3['Producto'].value_counts()
ventas_t4 = tienda4['Producto'].value_counts()
print("Tienda 1:")
print("Más vendido:", ventas_t1.idxmax(), "-", ventas_t1.max(), "ventas")
print("Menos vendido:", ventas_t1.idxmin(), "-", ventas_t1.min(), "ventas")

print("\nTienda 2:")
print("Más vendido:", ventas_t2.idxmax(), "-", ventas_t2.max(), "ventas")
print("Menos vendido:", ventas_t2.idxmin(), "-", ventas_t2.min(), "ventas")

print("\nTienda 3:")
print("Más vendido:", ventas_t3.idxmax(), "-", ventas_t3.max(), "ventas")
print("Menos vendido:", ventas_t3.idxmin(), "-", ventas_t3.min(), "ventas")

print("\nTienda 4:")
print("Más vendido:", ventas_t4.idxmax(), "-", ventas_t4.max(), "ventas")
print("Menos vendido:", ventas_t4.idxmin(), "-", ventas_t4.min(), "ventas")

```

```python
import seaborn as sns
def obtener_extremos(df, nombre_tienda):
    conteo = df['Producto'].value_counts()
    mas_vendido = conteo.idxmax()
    menos_vendido = conteo.idxmin()
    return [
        {'Producto': mas_vendido, 'Ventas': conteo.max(), 'Tienda': nombre_tienda, 'Tipo': 'Más vendido'},
        {'Producto': menos_vendido, 'Ventas': conteo.min(), 'Tienda': nombre_tienda, 'Tipo': 'Menos vendido'}
    ]

datos = []
datos += obtener_extremos(tienda, 'Tienda 1')
datos += obtener_extremos(tienda2, 'Tienda 2')
datos += obtener_extremos(tienda3, 'Tienda 3')
datos += obtener_extremos(tienda4, 'Tienda 4')

df = pd.DataFrame(datos, columns=['Tienda', 'Producto', 'Ventas', 'Tipo'])


plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Producto', y='Ventas', hue='Tienda', palette='Set2')

# Añadir etiquetas para diferenciar más/menos vendidos
for index, row in df.iterrows():
    plt.text(index, row['Ventas'] + 0.5, row['Tipo'], ha='center', va='bottom', fontsize=8)

plt.title('Productos Más y Menos Vendidos por Tienda')
plt.ylabel('Cantidad de Ventas')
plt.xlabel('Producto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
